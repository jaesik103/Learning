```python
# subplot : 하나의 페이지에 여러개의 그래프 그림
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)  # subplot(행의수, 열의수, 위치)
plt.plot(x1, y1, 'yo-')
plt.title(' A table of 2 subplot ')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time ')
plt.ylabel('Undamped ')
print(ax2)
plt.show()
```

```python
# 선 그래프
# 시도별 진출입 인구수
import pandas as pd

df = pd.read_excel('dataset/시도별 전출입 인구수.xlsx', header=0)
df.head(25)
```

```python
# 데이터 전처리 작업 
# 1. Nan 데이터 처리
#    전출지의 Nan 자료를 수정 -> 이전 자료의 값으로 대체
#    Nan 자료를 수정 : method = ffill 이전자료로 대체, bfill : 이후 자료로 대체
#    value = 값  -> 지정하는 값으로 수정
df.fillna(method='ffill', inplace=True)   

# 2. 필요한 자료만 추출
#    서울특별시에서 다른 지역으로 전출한 자료만 추출
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[ mask ]
# df_seoul = df[  (df['전출지별'] == '서울특별시') \
#                & (df['전입지별'] != '서울특별시')]

# 필요없는 컬럼 삭제 : '전출지별'은 같은 자료이므로 삭제
df_seoul = df_seoul.drop(['전출지별'], axis=1)

# 3. 컬럼명 또는 인덱스 재 설정
#    '전입지별'을 '전입지'로 변경
df_seoul = df_seoul.rename({'전입지별': '전입지'}, axis=1)

#    '전입지' 컬럼을 인덱스로 변경
df_seoul.set_index('전입지', inplace=True)
df_seoul
```

```python
# 남북한 전력 발전량
df = pd.read_excel('dataset/남북한발전전력량.xlsx', convert_float=True)
df.head(10)

# 북한의 자료만 가져 옴
df_1 = df.loc[5:9]
df_1.drop('전력량 (억㎾h)', axis=1, inplace=True)
df_1
```

```python
# 발전전력별 컬럼을 인덱스로 지정
df_1.columns
df_1.set_index('발전 전력별', inplace=True)
df_1
```

```python
df_2 = df_1.T
# 년도별로 증감율, '합계' 컬럼명을 '총발전량'으로 변경
df_2 = df_2.rename(columns={'합계' : '총발전량'})
df_2['총발전량 -1년'] = df_2['총발전량'].shift(1)
```

```python
# (증감율 총발전량/이전 년도의 발전량 -1)*100
df_2['증감율'] = ((df_2['총발전량']/df_2['총발전량 -1년'])-1)*100
df_2
```

```python
# 그래프 그리기 : 수력, 화력 -> 증감율
ax1 = df_2[['수력','화력']].plot(kind='bar', figsize=(20, 10), stacked=True)
ax2 = ax1.twinx()

ax2.plot(df_2.index, df_2['증감율'], ls='--', marker='o')
ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('발전량(억 Kwh)')
ax2.set_ylabel('전년 대비 증감율(%)')

plt.title('북한 전력 발전량')
ax1.legend(loc='best')

plt.show()
```
