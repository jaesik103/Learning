```python
# 그룹연산 : df.groupby(그룹을 만들고자 하는 열 또는 열 리스트)

df = titanic.loc[:, ['age', 'sex', 'class', 'fare','survived']]
df.head()
```

```python
df['class'].unique()
len(df)

# class를 기준으로 그룹 생성
class_group = df.groupby(['class'])
print(list(class_group))
```

```python
for key, group in class_group:
    print('key : ', key)
    print('group의 수 :', len(group))
    print(group.head())
```

```python
    
# 특정 그룹을 선택하고자 할 경우 : 그룹한 결과.get_group(그룹명)
print(np.floor(class_group.get_group('First')['age'].mean()))
```

```python
# class별 평균 나이와 평균 요금 출력
print(class_group.mean()[['age','fare']])
# 평균 나이가 가장 많은 클래스의 평균나이 출력
print(np.floor(np.max(class_group.mean()['age'])))
```

```python
# 클래스별 성별 기준으로 그룹
class_sex_group = df.groupby(['class', 'sex'])

# 가장 나이가 많은 클래스와 성, 적은 클래스와 성을 출력
age_max_list = {}
age_min_list = {}
for key, value in list(class_sex_group.age):
    age_max_list[key] = value.max()
    age_min_list[key] = value.min()

max_value = max(age_max_list.values())
min_value = min(age_min_list.values())

for min_key, max_key in zip(age_min_list, age_max_list):
    if min_value == age_min_list[min_key]:
        print('min : {}-{}'.format(min_key, min_value))
    if max_value == age_min_list[max_key]:
        print('min : {}-{}'.format(max_key, max_value))
```

```python
# 클래스별 성별로 그룹을 지어 생존자의 자료만 추출
# survived가 1인 경우 생존, 0이면 사망
def surfuction(value):
    if value == 1:
        value='생존'
    else:
        value='사망'
    return value


survived_serise = class_sex_group.survived
survived_serise.applymap(surfuction(survived_serise))
```

```python
# 열마다 다른 함수를 적용 : group.arr({'열1' : 함수1, '열2' : 함수2})
# age 는 평균값, 요금은 평균 가장 비싼요금, 가장 싼 요금
agg_sep = class_group.agg({'fare' : ['max', 'min'], 
                          'age' : 'mean'})
agg_sep
```

```python
# 그룹 연산 데이터 변환 : group.transform(매핑함수)
# 그룹별 age 열의 평균 집계 연산
age_mean = class_group.age.mean()
print(age_mean)
print('\n')

# 그룹별 age 열의 표준편차 집계 연산
age_std = class_group.age.std()
print(age_std)

# 그룹 객체의 age 열을 iteration으로 z-score를 계산하여 출력
for key, group in class_group.age:
    group_zxcore = (group-age_mean.loc[key])/age_std.loc[key]
    print(' --orgin : ', key)
    print(group_zxcore.head(3))
    print()
```

```python
# z-score를 계산하는 사용자 함수 정의
def z_score(x):
    return (x - x.mean()) / x.std()

age_zscore = class_group.age.transform(z_score)
age_zscore.loc[[1, 9, 0]]
```

```python
# 그룹 객체 필터링 : group.filter(조건식 함수)
# 데이터 개수가 200개 이상인 그룹만 데이터프레임으로 변환
filter_group = class_group.filter(lambda x : len(x) > 300)
filter_group.head()
len(filter_group)
```
