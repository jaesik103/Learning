```python
x1 =np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2*np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)  # (행의수 열의수 위치)
plt.plot(x1, y1, 'yo-')
plt.title('A table of 2 subplot')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time')
plt.ylabel('Undaped')
print(ax2)
plt.show()
```

```python
# 선 그래프
# 시도별 진출입 인구수
import pandas as pd
df = pd.read_excel('dataset/시도별 전출입 인구수.xlsx', header=0)
df
```

```python
# 전출지의 Nan 자료를 수정 -> 이전 자료의 값으로 대체
df.fillna(method='ffill', inplace=True)  # ffill : 이전자료로 대체, bfill: 이후 자료로 대체, value = 지정값으로 대체
df
```

```python
# 서울특별시에서 다른 지역으로 전출한 자료만 추출
mask = (df['전출지별']=='서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul.drop('전출지별', axis=1, inplace=True)
df_seoul.set_index('전입지별', inplace=True)
df_seoul
```

```python
# 서울에서 경기도로 전입한 인수를 그래프로 확인
sr_one = df_seoul.loc['경기도']
sr_one
```

```python
plt.figure(figsize=(20, 10))
plt.plot(sr_one.index, sr_one.values, marker='o', markersize=15)
plt.style.use('ggplot')
plt.title('서울 -> 경기')
plt.ylabel('명 수')
plt.xlabel('년도 별')
plt.xticks(rotation='vertical', size=10)
plt.legend(labels=['서울 -> 경기'], loc=0)
plt.show()
```

```python
# 전라남도, 충청남도, 경상남도, 경기도로 이동한 인구수
# 하나의 화면에 출력
df.fillna(value = 0, inplace=True)
mask = (df['전입지별'] == '전라남도')|(df['전입지별'] == '충청남도')|(df['전입지별'] == '경상남도')|(df['전입지별'] == '경기도')
df_seoul = df[mask]



# 전입지 칼럼을 인덱스로 변경
df_seoul.set_index('전입지별', inplace=True)

sr_one = df_seoul.loc['전라남도']

# 그래프
plt.plot(sr_one.index, sr_one.values)
plt.show()
```
