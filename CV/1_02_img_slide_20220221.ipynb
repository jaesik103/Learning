```python
import numpy as np
import cv2
import sys
```

```python
img = cv2.imread('./fig/puppy.bmp', 1)
print(img.shape)

h, w = img.shape[:2]
img_re = cv2.resize(img, (w//2, h//2), interpolation = cv2.INTER_AREA)

if img is None:
    sys.exit('imread failed')

cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image', img_re)

while True:
    key = cv2.waitKey()
    if key == ord('q'):
        break
        
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
```

```python
imgBGR = cv2.imread('./fig/cat.bmp')
imgGRAY = cv2.imread('./fig/cat.bmp', 0)

if imgBGR is None:
    sys.exit('imread failed')

imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

plt.subplot(131), plt.imshow(imgBGR), plt.axis('off')
plt.subplot(132), plt.imshow(imgRGB), plt.axis('off')
plt.subplot(133), plt.imshow(imgGRAY, cmap = 'gray'), plt.axis('off')
plt.show()
```

```python
## 이미지 슬라이드

import os

img_list = os.listdir('./fig/images')
#print(img_list)

img_files = []
for i in img_list:
    img_file = './fig/images/'+i
    #print(img_file)
    img_files.append(img_file)

cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('image', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    img = cv2.imread(img_files[idx])
    cv2.imshow('image', img)
    
    if cv2.waitKey(3000) == ord('q'):
        cv2.destroyAllWindows()
        break
        
    idx += 1
    if idx>=len(img_files):
        idx = 0
        
```
