# openCV
- cv2.imread 로 이미지를 읽어 저장함. 넘파이어레이로 저장하며, RGB가 아닌 BGR로 저장함
- cv2.namedWindow() : 윈도우창 이름. 마음대로 저장해도 됨
- cv2.imshow(윈도우이름, 이미지)
- cv2.waitKey() : 키 입력을 받음.
- cv2.destroyWindows() : 윈도우창 종료
- 높이, 너비 순으로 크기 나타남.
- cv2.resize(interpolation = ) : 어떤 방식으로 늘리거나 줄일것인지에 대한 플래그
- ord() : 아스키코드로 변환해주는 함수

```python
img = cv2.imread('./fig/cat.bmp', 1)

if img is None:
    sys.exit('imread failed')

cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image', img)

while True:
    key = cv2.waitKey()
    if key == ord('q'):
        break
    elif key == ord('s'):
        cv2.imwrite('./fig/key_test.png', img)
        print('image saved')


cv2.destroyAllWindows()
```

```python
import numpy as np
import sys
import cv2
import matplotlib.pyplot as plt
import platform

imgBGR = cv2.imread('./fig/snowman.jpg', 1)
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
imgGRAY = cv2.imread('./fig/snowman.jpg', 0)

if imgBGR is None:
    sys.exit('imread failed')

plt.figure(figsize=(6,12))
plt.subplot(131), plt.imshow(imgBGR), plt.axis('off')
plt.subplot(132), plt.imshow(imgRGB), plt.axis('off')
plt.subplot(133), plt.imshow(imgGRAY), plt.axis('off')
plt.show()
```

```python
img = cv2.imread('./fig/snowman.jpg', 1)
h, w = img.shape[:2]
img_resize = cv2.resize(img, (w//2, h//2), interpolation = cv2.INTER_LINEAR)

if img is None:
    sys.exit('imread failed')

cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.moveWindow('image', 200,200)
cv2.imshow('image', img_resize)

cv2.waitKey()
cv2.destroyAllWindows()
```
